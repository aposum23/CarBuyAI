```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset_path = 'car_data.csv'
cars_dataset = pd.read_csv(dataset_path)
print(cars_dataset)
```

```python
males = cars_dataset[cars_dataset['Gender'] == 'Male']
females = cars_dataset[cars_dataset['Gender'] == 'Female']
```

```python
males_buy = males[males['Purchased'] == 1]
males_not_buy = males[males['Purchased'] == 0]
females_buy = females[females['Purchased'] == 1]
females_not_buy = females[females['Purchased'] == 0]
```

```python
x_male_buy = males_buy['AnnualSalary']
y_male_buy = males_buy['Age']
x_male_not_buy = males_not_buy['AnnualSalary']
y_male_not_buy = males_not_buy['Age']
plt.plot(y_male_buy, x_male_buy, 'o', color='green')
plt.plot(y_male_not_buy, x_male_not_buy, 'o', color='red')
plt.ylabel('Male age')
plt.xlabel('Male salary')
plt.show()
```

```python
x_female_buy = females_buy['AnnualSalary']
y_female_buy = females_buy['Age']
x_female_not_buy = females_not_buy['AnnualSalary']
y_female_not_buy = females_not_buy['Age']
plt.plot(y_female_buy, x_female_buy, 'o', color='green')
plt.plot(y_female_not_buy, x_female_not_buy, 'o', color='red')
plt.ylabel('Female age')
plt.xlabel('Female salary')
plt.show()
```

```python
plt.plot(y_male_buy, x_male_buy, 'o', color='green')
plt.plot(y_male_not_buy, x_male_not_buy, 'o', color='red')
plt.plot(y_female_buy, x_female_buy, 'o', color='green')
plt.plot(y_female_not_buy, x_female_not_buy, 'o', color='red')
plt.show()
```

```python
w1 = -2.7
w2 = 300000
age = []
salary = []
for i in range(15, 60):
    age.append(i)
for i in range(15, 60):
    salary.append(w1 * i * 2000 + w2)
```

```python
plt.plot(y_male_buy, x_male_buy, 'o', color='green')
plt.plot(y_male_not_buy, x_male_not_buy, 'o', color='red')
plt.plot(y_female_buy, x_female_buy, 'o', color='green')
plt.plot(y_female_not_buy, x_female_not_buy, 'o', color='red')
plt.plot(age, salary)
plt.show()
```

```python
def predict(age, salary):
    return w1 * age * 2000 + w2 < salary
```

```python
count_right = 0
count = 0
for i, data in cars_dataset.iterrows():
    count += 1
    prediction = predict(data['Age'], data['AnnualSalary'])
    if prediction and data['Purchased'] == 1:
        count_right += 1
    elif not prediction and data['Purchased'] == 0:
        count_right += 1
```

```python
print(count_right)
```

```python
print(count)
```

```python
procent_right = count_right / count
print(f'Right predicted coef: {procent_right}')
```
